$\text{circ } \mathbf a = (\mathbf a_{(k-l) \text{ mod } N})_{k,l=0}^{N-1}$<br>

$\text{circ } \mathbf a = N^2 \overline{\mathcal{F}_N}\text{diag}(\mathbf{\hat a})\mathcal{F}_N$<br>

$(\text{diag}(\mathbf{\hat a})\mathcal{F}_N))_{lj} = \sum_{k=0}^{N-1}\text{diag}(\mathbf{\hat a})_{lk}(\mathcal{F}_N)_{kj}$<br>

### g)

We can write the function to be interpolated as $g(x):=\sum_{j=0}^{N-1}s_jI_N(x-\frac{j}{N})$, s.t. $\hat{\mathbf{g}}=\mathcal{F}\mathbf{s}$, where $\mathbf{s}$ are the given values to be interpolated.
By $\mathbf{f)}$ we can easily calculate our coefficients $\mathbf{a} = (a_0,\ldots,a_{N-1})^T$ thusly:
$\mathbf{a}=N\overline{\mathcal{F}_N}\hat{\mathbf{a}}=N\overline{\mathcal{F}_N}\frac{\hat{\mathbf{g}}}{\hat{\mathbf{Nf}}}=\overline{\mathcal{F}_N}\frac{\hat{\mathbf{g}}}{\hat{\mathbf{f}}}$, as we know that $\hat{f}_h\neq0$

```python
import numpy as np
import matplotlib.pyplot as plt

def DFT(N):
    """
    returns the Discrete Fourier Transform matrix of size N
    """
    dft = np.zeros((N,N),dtype=complex)
    for l in range(N):
        for k in range(l,N):
            val = np.exp(1j*(-2*np.pi*l*k/N))
            dft[l,k] = val
            dft[k,l] = val
    return dft/N

def interpolate(s,ckf):
    """
    interpolates a function with values s in our interval, 
        using the function given by ckf
    returns the weights a we have to give the shifts of our interpolating function
    """
    N = len(s)
    dft = DFT(N)
    ghatk = dft @ s
    n = len(ckf)
    if n==N:
        fhatk = ckf
    else:
        midpt = np.floor(n/2)
        fhatk = np.zeros_like(ghatk)
        for i in range(n):
            fhatk[int(i%N)] += ckf[int((midpt+i)%n)]
    # Adding a failsafe for very small factors ckf
    return np.real(np.conjugate(dft) @ np.divide(ghatk, fhatk, out=np.zeros_like(ghatk), \
            where= np.abs(fhatk) > 1e-14))

def getfnc(ckf,x):
    """
    ckf: complex Fourier coefficients of a function (centered)
    x: points where the function is to be evaluated
    returns the function defined by ckf evaluated at x
    """
    N = len(ckf)
    f = np.sum([ckf[k]*np.exp((2*np.pi*(k-np.floor(N/2))*x)*1j) \
        for k in range(N)],axis=0,dtype=complex)
    return f

def superpos(a, f):
    """
    a: weights of the function shifted with j/N for a[j]
    f: function to be shifted and superpositioned (passed as y-values)
    returns superposition of the shifts of the function with weigths a
    """
    func = np.zeros_like(f, dtype=complex)
    for i in range(len(a)):
        func += a[i]*np.roll(f,-int(i*len(f)/len(a)))
    return func
```

$a_h=(N\overline{\mathcal{F}_N}\mathbf{\hat{a}})_h=\frac1{N}\sum_{j=0}^{2n}e^{-2\pi ihj/(2n+1)}=\delta_{h0}$ 

Therefore, when we interpolate the fundamental interpolant with the Dirichlet kernel we simply get back the Dirichlet kernel, so we will only be plotting $D_n$ in the following plot. We can confirm this by printing out $\mathbf a$.

For ease of plotting we will plot the closed form of the Dirichlet Kernel:
$\frac1{2n+1}\sum_{k=-n}^n e^{2\pi ikx}=\frac{e^{-2\pi inx}}{2n+1} \sum_{k=0}^{2n}e^{2\pi ikx}=\frac{e^{-2\pi inx}}{2n+1}\frac{1-e^{2\pi i(2n+1)x}}{1-e^{2\pi ix}} = \frac{e^{-2\pi inx}}{2n+1} \frac{e^{2\pi i(2n+1/2)x}-e^{-\pi ix}}{e^{\pi ix}-e^{-\pi ix}}$
$= \frac{1}{2n+1} \frac{e^{2\pi i(n+1/2)x}-e^{-2\pi i(n+1/2)x}}{e^{\pi ix}-e^{-\pi ix}} = \frac1{2n+1} \frac{sin(2\pi (n+1/2)x)}{sin(\pi x)}$

```python
def dirichlet(x,n):
    numerator = np.sin(np.pi*(2*n+1)*x) 
    denominator = (2*n+1)*np.sin(np.pi*x)
    return np.divide(numerator, denominator, \
        out=np.full_like(denominator,1), where= np.abs(denominator) > 1e-12)
```

$a_h=(N\overline{\mathcal{F}_N}\mathbf{\hat{a}})_h=\sum_{j=0}^{2n-1}\hat{a}_je^{-2\pi ihj/(2n)}=-\frac{2n+1}{8n^2}e^{-2\pi ihn/(2n)}+\frac{2n+1}{4n^2}\sum_{j=0}^{2n-1}e^{-2\pi ihj/(2n)}\Rightarrow$ 
$a_h=\begin{cases} -\frac{2n+1}{8n^2}e^{-2\pi ihn/(2n)}+\frac{2n+1}{2n} & \enspace h = 0 \\ -\frac{2n+1}{8n^2}e^{-2\pi ihn/(2n)} & \enspace h = 1,\ldots,2n-1 \end{cases}$

Where we get the last equality from $\mathbf{1f)}$\
Now,
$I_{2n}(x)=\sum_{j=0}^{2n-1}a_jD_n(x-\frac{j}{2n})=\frac{2n+1}{2n}D_n(x)-\frac{2n+1}{8n^2}\sum_{j=0}^{2n-1}D_n(x-\frac{j}{2n})e^{-2\pi ijn/2n}$\
$ \quad\qquad = \frac{1}{2n}\sum_{k=-n}^{n}e^{2\pi ikx}(1-\frac{1}{4n}\sum_{j=0}^{2n-1}e^{-2\pi i(k+n)j/2n})$\
$ \quad\qquad = \frac1{2n}\sum_{k=-n}^ne^{2\pi ikx}-\frac1{4n}(e^{2\pi inx}+e^{-2\pi inx})$\
As $(k+n) \text{ mod } 2n \equiv0$ only if $k=\pm n$ for $k=-n,\ldots,n$\
$ \quad\qquad = \frac1{2n}\sum_{k=-(n-1)}^{n-1}e^{2\pi ikx}+\frac1{4n}(e^{2\pi inx}+e^{-2\pi inx})$

With the modified Dirichlet kernel we halve the highest frequency terms, and need one less datapoint. We still in essence use the same number of interpolation points as "normal" Dirichlet kernel, as now the first and last point of each period is a root of the modified kernel. This also has the side effect of better approximating the fundamental interpolant, as you will see in the plot. Using fewer points stretches them out so we will now see that our points are slightly further apart than in $\mathbf{a)}$.

Now we want to write the modified dirichlet kernel in closed form. Using $\mathbf{3a)}$ we get:
$\frac1{2n}\sum_{k=-(n-1)}^{n-1}e^{2\pi ikx}+\frac1{4n}(e^{2\pi inx}+e^{-2\pi inx}) = \frac1{2n}(\frac{e^{2\pi i(n-1/2)x}-e^{-2\pi i(n-1/2)x}}{e^{\pi ix}-e^{-\pi ix}}+\frac12(e^{2\pi inx}+e^{2\pi inx}))$ $\qquad=\frac1{4n}\frac{sin(2\pi (n+\frac12)x)+sin(2\pi (n-\frac12)x)}{sin(\pi x)} = \frac1{2n}\frac{sin(2\pi nx)cos(\pi x)}{sin(\pi x)} = \frac1{2n}\frac{sin(2\pi nx)}{tan(\pi x)}$

Where we get the first equality by expanding and the second from trigonometric identities. 

```python
def m_dirichlet(x,n):
    numerator = np.sin(2*np.pi*n*x)
    denominator = 2*n*np.tan(np.pi*x)
    return np.divide(numerator, denominator, \
        out=np.full_like(denominator,1), where= np.abs(denominator) > 1e-12)
```

```python
n = [5, 10, 17]
x = np.linspace(-1/2,1/2,1000)

for ns in n:
    y = m_dirichlet(x, ns)
    roots=np.array([i/(2*ns) for i in range(-ns,ns+1) if i != 0])
    plt.scatter(roots, np.zeros_like(roots),color="black", s=6)
    plt.plot(x,y,label="n = {0}".format(ns))
plt.scatter(0,1,color="black", s=6)
plt.title("Modified Dirichlet kernel for several n")
plt.legend()
plt.show()
```

```python
def wavelet_space(x,n):
    """
    Takes in n/2 as n
    """
    N = int(2*n)
    n = int(n)
    fnc = np.zeros_like(x)
    for i in range(n,N):
        fnc += np.cos(2*np.pi*(x-1/(2*N))*i)
    fnc *= 2/(N+1)
    fnc -= np.cos(2*np.pi*N*x)/(n+1)
    return fnc
```

```python
from PIL import ImageOps, Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import fft
```

```python
def cart_to_polar(complex_array):
    """Calculates polar coordinates for elements in complex input array, returned in separate real matrices"""
    return abs(complex_array), np.angle(complex_array)

def polar_to_cart(amplitude, angle):
    """From same-dimension amplitude- and angle coordinates for complex numbers, return cartesian representation in complex matrix"""
    return amplitude * np.exp(1j * angle)
```

There is a clear grid pattern stemming from the dot half-toning here. For removing the dots, but keeping all other patterns in a dithered image, removing these spikes would be a good idea. Beneath, a high frequency filter (as explained here: https://www.mathematica-journal.com/2006/09/07/dither-removal/?fbclid=IwAR1bVz3Ui743ztZmJnShmQalyJB6qJsXEjKLgjKbXN5z2XHCjE2Q4Wndi0k) is implemented.

In short: as most of the information we want to keep are at low frequencies, but highly periodic features (like the dots) also can be in higher frequencies (as seen in the figure. Lower frequencies are in the corner, as $F_{ij}$ is $N_1$-periodic for $i$ and $N_2$-periodic for $j$), we want to keep the low frequency coefficients but limit the higher. To accomplish this, a gauusian curve matrix ($1$ at center but quickly goes toward $0$ in all directions$) is generated and centered at the lowest frequencies. This is then multiplied element-wise with the amplitude matrix. Note that this is in agreement with the discovery in last sub-task, that changing the amplitudes of discrete Fourier coefficients will not majorly disturb the data.

```python
def A_filter(A, var_frac):
    """Filter out high frequencies on a 2 dimensional discrete fourier coef grid"""
    A_mask = np.zeros_like(A)
    rows, cols = np.shape(A)
    for i in range(rows):
        for j in range(cols):
            A_mask[i,j] = np.exp(-(j - cols/2)**2/(var_frac * cols)**2) * np.exp(-(i - rows/2)**2/(var_frac * rows)**2)
    A_mask = np.roll(A_mask, rows//2, axis=0)
    A_mask = np.roll(A_mask, cols//2, axis=1)
    return A * A_mask
```

Consider a function $F : \mathbb{T} \times \mathbb{T} \rightarrow \mathbb{C}$. Let's assume that $F(\textbf{x}) = f^{(1)}(x_1)f^{(2)}(x_2)$, $\enspace f^{(1)}, f^{(2)} : \mathbb{R}^+\rightarrow \mathbb{C}$. The discretization $[F([\frac{i}{N_1}, \frac{j}{N_2}])]_{(i, j) = (0,0)}^{(N_1-1,N_2-1)}$ can then be written $[f^{(1)}(\frac{i}{N_1})f^{(2)}(\frac{j}{N_2})]_{(i, j) = (0,0)}^{(N_1-1,N_2-1)} = [f_i^{(1)}f_j^{(2)}]_{(i, j) = (0,0)}^{(N_1-1,N_2-1)} = \textbf{f}^{(1)}\textbf{f}^{(2)^T}$

Then the separate vectors $\textbf{f}^{(k)}$ can be shifted, $N_1$ times for $\textbf{f}^{(1)}$, $N_2$ times for $\textbf{f}^{(2)}$. It thus makes sence to define $V_{N, F} = \text{span}\{\tau_{N_1}^{i}f^{(1)}\tau_{N_2}^j f^{(2)}\vert i, n \in \mathbb{N}\}$ For the discreet case, this corresponds to shifting each of the vectors in the tensor product. This again corresponds to shifting the entire $F$ function horizontally ($f^{(1)}$) and vertically ($f^{(2)}$), i.e. moving a picture without distortion around, looping at the edges.
